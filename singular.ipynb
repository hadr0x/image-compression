```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
plt.rcParams['figure.figsize'] = [16, 8]

A = imread('img/dog.jpg')
X = np.mean(A,-1) # Passaggio da RGB a scala di grigi


img = plt.imshow(X)
plt.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=False)

S = np.diag(S)

for r in (5, 20, 100):
    # Calcolo le immagini approssimate di rango 5, 20, 100
    Xapprox = U[:,:r] @ S[0:r,:r] @ VT[:r,:]
   

    img = plt.imshow(Xapprox)
    plt.set_cmap('gray')
    plt.axis('off')
    plt.title(f"r = {r}")
    plt.show()
    
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values', fontdict={'fontsize' : 20}, pad=20)
plt.xticks(range(0,2300,200))
plt.show()
```

```python
plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title('Singular Values: Cumulative Sum', fontdict={'fontsize' : 20}, pad=20)
plt.xticks(range(0,2300,200))
plt.show()
```
